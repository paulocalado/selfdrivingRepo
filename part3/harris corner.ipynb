```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
chess = cv2.imread("chessboard.jpg")
cv2.imshow("chess",chess)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
gray_image = cv2.cvtColor(chess,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.cornerHarris(gray_image,5,3,0.1)
plt.imshow(corners,cmap="gray")
```

```python
truck_img = cv2.imread("Truckrr.png")
```

```python
gray_truck = cv2.cvtColor(truck_img,cv2.COLOR_BGR2GRAY)
```

```python
truck_corners = cv2.cornerHarris(gray_truck,7,5,0.1)
corners_dilate = cv2.dilate(truck_corners,np.ones((8,8),np.uint8),iterations=1)
plt.imshow(truck_corners,cmap="gray")
```

```python
plt.imshow(corners_dilate,cmap="gray")
```

```python
corners_dilate.shape
```

```python
corners_dilate.max()
```

```python
image_color_corners = truck_img.copy()
image_color_corners[corners_dilate>0.03*corners_dilate.max()] = [0,255,0]

cv2.imshow("corners",image_color_corners)
cv2.waitKey()
cv2.destroyAllWindows()
```
