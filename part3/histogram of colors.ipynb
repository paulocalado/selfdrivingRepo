```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
image = cv2.imread("Tesla.jpg")
```

```python
image.shape
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histogram = cv2.calcHist([image],[i],None,[256],[0,256])
    plt.plot(histogram,color=col)
    plt.xlim([0,256])
plt.show()
```
