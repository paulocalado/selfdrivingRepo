```python
#### split channels
B,G,R = cv2.split(image)
B.shape
```

```python
G.shape
```

```python
cv2.imshow("Blue Channel",B)
cv2.waitKey(0)
cv2.destroyAllWindows()
#OpenCV interprets as a gray scale becaus it's a single channel
```

```python
####create our own blue channel

zeros= np.zeros(image.shape[:2],dtype="uint8")

cv2.imshow("Now the blue channel",cv2.merge([B,zeros,zeros]))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##### mergin all three channels
merged_image = cv2.merge([B,G,R])
cv2.imshow("Merged Images",merged_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
greener_img = cv2.merge([B,G+100,R])
cv2.imshow("Greener Image", greener_img)
cv2.waitKey()
cv2.destroyAllWindows()
```
